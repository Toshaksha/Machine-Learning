# ***Diabetes Prediction using Support Vector Machine (SVM)***

### Objective:
This machine learning project uses patient health data to predict whether a person is diabetic or not
based on features like glucose level, BMI, age, etc.

### Workflow Overview
- ***Import Libraries*** – Load essential packages
- ***Load & Explore Data*** – Understand structure and statistics
- ***Separate Features & Labels*** – Define input features (X) and target label (Y).
- ***Train-Test Split*** – Prepare for training
- ***Data Standardization*** –
Standardize features to improve model performance.
- ***Model Building*** – Train an SVM classifier
- ***Evaluate Model*** – Check accuracy
- ***Make Predictions*** – Test on new input
### Problem Type:
- Binary Classification (0 = Non-Diabetic, 1 = Diabetic)

### Dataset:
- Source: Pima Indian Diabetes Dataset
- Rows: 768 samples
- Features: 8 numerical columns (Pregnancies, Glucose, BloodPressure, etc.)
- Label: Outcome (0 or 1)



### Step 1: Import Dependencies
We'll use NumPy and Pandas for data handling, and Scikit-learn for modeling and evaluation.



```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Step 2: Load and Explore the Dataset
Load the CSV data, check its structure, and understand the label distribution.

```python
# Load the diabetes dataset into a Pandas DataFrame
diabetes_data = pd.read_csv("/content/diabetes.csv")
```

```python
# Display first 5 rows
diabetes_data.head()
```

```python
# Check dataset shape: (rows, columns)
diabetes_data.shape
```

```python
# Get statistical summary of the data
diabetes_data.describe()
```

```python
# Check target value counts
diabetes_data['Outcome'].value_counts()
```

**Label Explanation**
- 0 → Non-Diabetic
- 1 → Diabetic

```python
# Compare feature means grouped by target class
diabetes_data.groupby('Outcome').mean()
```

This helps identify which features differ between diabetic and non-diabetic patients (e.g., glucose, BMI).

### Step 3: Separate Features and Labels
Separate the input variables (X) from the target variable (Y).

```python
# X contains all features except the target 'Outcome'
X = diabetes_data.drop(columns= 'Outcome', axis = 1)
```

```python
# Y contains the labels (0 or 1)
Y = diabetes_data['Outcome']
```

```python
print("Feature shape:", X.shape)
print("Label shape:", Y.shape)
```

### Step 4: Train-Test Split
Split the dataset into training (80%) and testing (20%) sets, using stratification to maintain class balance.

```python
X_train , X_test , Y_Train , Y_test = train_test_split(X, Y , test_size= 0.2 , random_state=2, stratify=Y)
```

```python
print("Original dataset:", X.shape)
print("Training set:", X_train.shape)
print("Test set:", X_test.shape)
```

**Why Stratify?**
- Stratification ensures both train and test sets have the same proportion of diabetic and non-diabetic samples.

### Step 5: Data Standardization
Standardize the feature values to bring them onto a similar scale.

```python
# Initialize the scaler
scalar = StandardScaler()
```

```python
# Fit and transform the training data
X_train_scaled = scalar.fit_transform(X_train)
```

```python
print(X_train_scaled)
```

```python
# Only transform the test data (no fitting)
X_test_scaled = scalar.transform(X_test)
```

```python
print(X_test_scaled)
```

**Why Scale?**
- SVMs are sensitive to the magnitude of features. Standardization improves performance by ensuring all features contribute equally.

### Step 6: Train the Support Vector Machine Model
We’ll use a linear SVM since the classes in this dataset are fairly separable.

```python
# Create SVM classifier with linear kernel
classifier = svm.SVC(kernel = "linear")
```

```python
# Train the classifier on standardized training data
classifier.fit(X_train_scaled, Y_Train)
```

**Why SVM?**
- SVMs work well for small to medium-sized datasets and handle binary classification problems effectively.

### Step 7: Evaluate the Model
Evaluate performance on both the training and testing datasets using accuracy score.

```python
# Predict on training data
X_train_prediction = classifier.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_Train)
```

```python
print("Training Accuracy:", training_data_accuracy)
```

```python
# Predict on test data
X_test_prediction = classifier.predict(X_test_scaled)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Test Accuracy:", test_data_accuracy)
```

**Why Evaluate Both?**
- High training accuracy only = possible overfitting
- High test accuracy = good generalization
- Large gap between both = model might be under/overfitting

### Step 8: Predict for New Input
Make predictions using real input data (simulated patient details).

```python
# Example patient data (Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigree, Age)
input_data = (4,110,92,0,0,37.6,0.191,30)

# Convert input to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape to match model input
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# Standardize input data using the same scaler
std_data = scalar.transform(input_data_reshape)

# Make prediction
prediction = classifier.predict(std_data)
print(prediction)

# Interpret result
if (prediction[0] == 0):
  print("Person is Non -Diabetic.")
else:
  print("Person is Diabetic.")
```
